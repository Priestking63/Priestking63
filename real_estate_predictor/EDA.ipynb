```python
mask = df['Площадь кухни'].notna() & df['Общая площадь'].notna()
percentage = (df.loc[mask, 'Площадь кухни'] / df.loc[mask, 'Общая площадь']).mean()

print(f"Средний процент: {percentage:.2%}")

# Заполняем пропуски
mask_na = df['Площадь кухни'].isna()
df.loc[mask_na, 'Площадь кухни'] = (df.loc[mask_na, 'Общая площадь'] * percentage).round(1)
```

```python
s_mean = df[df['Количество комнат'] == 'студия']['Общая площадь'].mean()
room_area_ratio = (df['Общая площадь'] - df['Площадь кухни']) / s_mean

mask = df['Количество комнат'] == 'свободная планировка'

df.loc[mask, 'Количество комнат'] = np.floor(room_area_ratio[mask]).astype(int)
```

```python
def smart_floor_feature(floor, total_floors):
    
    if total_floors == 1:
        return 0.5
    relative = (floor - 1) / (total_floors - 1)
    
    if total_floors <= 5:  # хрущевки и малоэтажки
        if floor == 1:
            return relative * 0.7  # штраф за первый этаж
        elif floor == total_floors:
            return relative * 1.1  # небольшой бонус за последний
        else:
            return relative
            
    else:  # высотки
        if floor == 1:
            return relative * 0.6  # большой штраф за первый
        elif floor == total_floors:
            return relative * 1.4  # большой бонус за пентхаус
        elif floor >= total_floors - 2:
            return relative * 1.2  # бонус за высокие этажи
        else:
            return relative

df['умный_относ_этаж'] = df.apply(lambda x: smart_floor_feature(x['Этаж'], x['Этажей в доме']), axis=1)
```

```python
df['Год постройки'] = df.groupby('Тип дома')['Год постройки'].transform(
    lambda x: x.fillna(x.mean())
).round(0)
```

```python
def fill_na_smart(df, col, floor_col='Этажей в доме', group_col='Тип дома'):
    """Заполняет пропуски: 0 если <=5 этажей, иначе модой по группе"""
    mask_na = df[col].isna()
    mask_low = df[floor_col] <= 5
    
    df.loc[mask_na & mask_low, col] = 0
    df.loc[mask_na & ~mask_low, col] = df.loc[mask_na & ~mask_low, group_col].map(
        df.groupby(group_col)[col].agg(lambda x: x.mode()[0] if not x.mode().empty else 0)
    )
    return df
```

```python
df = fill_na_smart(df, 'Пассажирский лифт')
df = fill_na_smart(df, 'Грузовой лифт')
df[['Пассажирский лифт', 'Грузовой лифт']] = df[['Пассажирский лифт', 'Грузовой лифт']].replace('нет', 0)


```

```python
def process_year_built(df, year_col='Год постройки'):
    """Оптимальная обработка года постройки для небольшого датасета"""
    
    df['возраст_дома'] = 2025 - df[year_col]
    
    df['период_строительства'] = pd.cut(
        df[year_col], 
        bins=[0, 1960, 1980, 2000, 2015, 2025],
        labels=['старые', 'советские', 'постсовет', 'modern', 'новые']
    )
    
    df['возраст_бинар'] = (df['возраст_дома'] > 30).astype(int)  # старый/новый
    
    return df

# Применяем
df = process_year_built(df)
df = df.drop('Год постройки', axis = 1)
```

```python
import pickle


class SafeCategoricalEncoder:
    def __init__(self):
        self.encoding_maps = {}
        self.fitted_columns = []
    
    def fit(self, df, categorical_cols, target):
        self.encoding_maps = {}
        self.fitted_columns = categorical_cols
        
        for col in categorical_cols:
            n_categories = df[col].nunique()
            
            if n_categories <= 5:
                # Ordered Label Encoding
                means = df.groupby(col)[target].mean().sort_values()
                self.encoding_maps[col] = {
                    'type': 'ordered_label',
                    'mapping': {cat: i for i, cat in enumerate(means.index)}
                }
            else:
                # Bayesian Target Encoding
                global_mean = df[target].mean()
                group_means = df.groupby(col)[target].mean()
                group_counts = df.groupby(col)[target].count()
                
                self.encoding_maps[col] = {
                    'type': 'bayesian',
                    'global_mean': global_mean,
                    'group_means': group_means.to_dict(),
                    'group_counts': group_counts.to_dict()
                }
        return self
    
    def transform(self, df):
        df_encoded = df.copy()
        
        for col, encoding_info in self.encoding_maps.items():
            if col in df.columns:
                if encoding_info['type'] == 'ordered_label':
                    df_encoded[col] = df[col].map(encoding_info['mapping'])
                else:
                    # Bayesian encoding
                    global_mean = encoding_info['global_mean']
                    group_means = encoding_info['group_means']
                    group_counts = encoding_info['group_counts']
                    
                    def encode_value(x):
                        if x in group_means:
                            count = group_counts.get(x, 0)
                            return (group_means[x] * count + global_mean * 50) / (count + 50)
                        return global_mean
                    
                    df_encoded[col] = df[col].apply(encode_value)
        
        return df_encoded
    
    def fit_transform(self, df, categorical_cols, target):
        return self.fit(df, categorical_cols, target).transform(df)
    
    def save(self, path):
        with open(path, 'wb') as f:
            pickle.dump(self, f)
    
    @classmethod
    def load(cls, path):
        with open(path, 'rb') as f:
            return pickle.load(f)
```

```python
categorical_cols = df.drop('Ссылка', axis = 1).select_dtypes(include=['object', 'category']).columns.tolist()
encoder = SafeCategoricalEncoder()
df_enc = encoder.fit_transform(df, categorical_cols, 'Цена')
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
from statsmodels.tools.tools import add_constant

# VIF для проверки мультиколлинеарности
def calculate_vif(df, numerical_cols):
    X = add_constant(df[numerical_cols].dropna())
    vif_data = pd.DataFrame()
    vif_data["feature"] = X.columns
    vif_data["VIF"] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
    return vif_data

vif_results = calculate_vif(df_enc.drop('Ссылка', axis = 1), numerical_cols)
print("VIF факторы:")
print(vif_results.sort_values('VIF', ascending=False))
```

```python
def generate_eda_report(df, target_col='Цена'):
    """Генерирует комплексный отчет по данным"""
    
    print("=" * 50)
    print("КОМПЛЕКСНЫЙ АНАЛИЗ ДАННЫХ")
    print("=" * 50)
    
    # Базовая информация
    print(f"Размер датасета: {df.shape}")
    print(f"Пропуски: {df.isna().sum().sum()}")
    print(f"Дубликаты: {df.duplicated().sum()}")
    
    # Анализ целевой переменной
    print(f"\nЦЕЛЕВАЯ ПЕРЕМЕННАЯ ({target_col}):")
    print(f"Мин: {df[target_col].min():.2f}")
    print(f"Макс: {df[target_col].max():.2f}")
    print(f"Среднее: {df[target_col].mean():.2f}")
    print(f"Медиана: {df[target_col].median():.2f}")
    
    # Корреляции с целевой переменной
    numerical_cols = df.select_dtypes(include=['int64', 'float64']).columns
    if target_col in numerical_cols:
        correlations = df[numerical_cols].corr()[target_col].sort_values(ascending=False)
        print(f"\nТОП-10 корреляций с {target_col}:")
        print(correlations.head(10))
    
    # Категориальные переменные
    categorical_cols = df.select_dtypes(include=['object', 'category']).columns
    print(f"\nКАТЕГОРИАЛЬНЫЕ ПЕРЕМЕННЫЕ ({len(categorical_cols)}):")
    for col in categorical_cols:
        print(f"{col}: {df[col].nunique()} уникальных значений")

# Генерируем отчет
generate_eda_report(df)
```
